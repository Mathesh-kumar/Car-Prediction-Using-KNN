## **Machine learning model to predict whether the customer will buy a new car or not using k nearest neighbors**
### **Steps:**
  1. Importing needed libraries
  2. Data Preprocessing
  3. Training KNN model
  4. Predicting new results
  5. Predicting test results
  6. Making confusion matrix

```
# 1. importing needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
# 2. data preprocessing
# 2.1. importing needed dataset
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```
# 2.2. splitting dependent & independent variable
X = df.iloc[:, 2:-1].values
y = df.iloc[:, -1].values
```

```
# 2.3. splitting train & test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```
# 2.4. feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
# 3. training knn model
# since we need find euclidean distance between poins, use metric as minkowski
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn.fit(X_train, y_train)
```

```
# 4. predicting new results
knn.predict(sc.transform([[30, 87000]]))
```

```
# 5. predicting test results
y_pred = knn.predict(X_test)
pd.concat([pd.DataFrame(y_test, columns=['y_test']),pd.DataFrame(y_pred, columns=['y_pred'])], axis=1)
```

```
# 6. making confusion matrix and accuracy
from sklearn.metrics import confusion_matrix, accuracy_score
conMatrix = confusion_matrix(y_test, y_pred)
accScore = accuracy_score(y_test, y_pred)
print(conMatrix, accScore)
```
